```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python

file=pd.read_csv("http://users.stat.ufl.edu/~winner/data/airq402.dat", sep='\s+'  , engine='python')
```

```python
file.head()
```

```python
file.columns = ['City1', 'City2', 'Average Flight Fare', 'Distance','Average Weekly Passengers','Market Leading Airline (MLA)',
               'Market Share MLA', 'Average MLA Fare', 'Low Price Airline (LPA)', 'Market Share LPA', 'Average LPA Fare']
```

```python
file.info()
```

```python
file.describe()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.distplot(file['Average Flight Fare'])
plt.subplot(1,3,2)
sns.distplot(file['Distance'],color='green')
plt.subplot(1,3,3)
sns.distplot(file['Average Weekly Passengers'],color='red')
plt.figure(figsize=(20,6))
plt.subplot(1,3,1)
sns.boxplot(file['Average Flight Fare'],orient='vertical')
plt.subplot(1,3,2)
sns.boxplot(file['Distance'],color='green',orient='vertical')
plt.subplot(1,3,3)
sns.boxplot(file['Average Weekly Passengers'],color='red',orient='vertical')
```

```python
#Market Share MLA	Average MLA Fare	Market Share LPA	Average LPA Fare

plt.figure(figsize=(20,5))
plt.subplot(1,4,1)
sns.distplot(file['Market Share MLA'])
plt.subplot(1,4,2)
sns.distplot(file['Average MLA Fare'])
plt.subplot(1,4,3)
sns.distplot(file['Market Share LPA'])
plt.subplot(1,4,4)
sns.distplot(file['Average LPA Fare'])
plt.figure(figsize=(20,5))
plt.subplot(1,4,1)
sns.boxplot(file['Market Share MLA'])
plt.subplot(1,4,2)
sns.boxplot(file['Average MLA Fare'])
plt.subplot(1,4,3)
sns.boxplot(file['Market Share LPA'])
plt.subplot(1,4,4)
sns.boxplot(file['Average LPA Fare'])
```
